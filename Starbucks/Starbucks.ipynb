```python
# define useful function to create scatterplots of Calories against desired columns
def scatterplots(columns, ncol=None, figsize=(15, 8)):
    if ncol is None:
        ncol = len(columns)
    nrow = int(np.ceil(len(columns) / ncol))
    fig, axes = plt.subplots(nrow, ncol, figsize=figsize, squeeze=False)
    fig.subplots_adjust(wspace=0.5, hspace=0.6)
    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        ax.scatter(x = col, y = 'Calories', data=df, alpha=0.5)
        ax.set(xlabel=col, ylabel='Calories')
    nsubplots = nrow * ncol    
    for empty in range(i+1, nsubplots):
        axes.flatten()[empty].set_visible(False)
```

```python
features = [x for x in df.columns if x not in ['Beverage_category', 'Beverage', 'Beverage_prep']]
scatterplots(features, ncol=4, figsize=(15, 15))
```

Further, scatter plot also shows sugar , cholesterol, Total cabohydrates show strong relationship with calories. 
Protein, trans fat, total fat shows medium relationship with calories. 

Lets find which beverages have more sugar,Total cabohydrates and cholesterol.

```python
# Calories vs Sugars (g)
plt.figure(figsize=(15, 8))
sns.scatterplot(data=df, x="Calories", y="Sugars (g)",hue="Beverage_category")
plt.legend(loc = 'upper left')
plt.show()
```

```python
bev_sugar=pd.DataFrame(df.groupby('Beverage')['Sugars (g)'].mean())
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(y=bev_sugar.index, x=bev_sugar['Sugars (g)'], data=bev_sugar, ci=None, color="gray")
plt.xticks(rotation=90)
plt.title("Beverage Average Sugar")
plt.show()
```

```python
# Calories vs Cholesterol
plt.figure(figsize=(15, 8))
sns.scatterplot(data=df, x="Calories", y="Cholesterol (mg)",hue="Beverage_category")
plt.legend(loc = 'upper left')
plt.show()
```

```python
bev_chol=pd.DataFrame(df.groupby('Beverage')['Cholesterol (mg)'].mean())

plt.figure(figsize=(15, 8))
sns.barplot(y=bev_chol.index, x=bev_chol['Cholesterol (mg)'], data=bev_chol, ci=None, color="gray")
plt.xticks(rotation=90)
plt.title("Beverage Average Cholesterol")
plt.show()
```

```python
# Calories vs Total Carbohydrates
plt.figure(figsize=(15, 8))
sns.scatterplot(data=df, x="Calories", y="Total Carbohydrates (g)",hue="Beverage_category")
plt.legend(loc = 'upper left')
plt.show()
```

```python
bev_carb=pd.DataFrame(df.groupby('Beverage')['Total Carbohydrates (g)'].mean())

plt.figure(figsize=(15, 8))
sns.barplot(y=bev_carb.index, x=bev_carb['Total Carbohydrates (g)'], data=bev_carb, ci=None, color="gray")
plt.xticks(rotation=90)
plt.title("Beverage Average Carbohydrates")
plt.show()
```

Java Chip Beverage shows highest Sugar, Cholesterol and Carbohydrates.
